```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/titanic_train.csv')
df.head()
```

**Null Values**
* dropna() - used to drop a row with null values
* drop() - used to drop any row or column
* fillna() - to fill with a defined value at missing spaces

**Example of dropna()**
****

```python
details = {'name' : ['Abhay' , 'Vijay' , 'John' , 'Jacob' , 'Jerry'],
           'Salary' : ['$3000' , '$2545' , '$2000' , '$40000' , '$50000'],
           'exp' : [2.4 , 4.3, 5.5, np.nan , 7.9]}
d1 = pd.DataFrame(details)
d1
```

```python
d2 = d1.dropna(axis='rows')
d2
```

```python
d3 = d1.dropna(axis='columns')
d3
```

****

**Example of drop() function**
****

```python
details = {'name' : ['Abhay' , 'Vijay' , 'John' , 'Jacob' , 'Jerry'],
           'Salary' : ['$3000' , '$2545' , '$2000' , '$40000' , '$50000'],
           'exp' : [2.4 , 4.3, 5.5, 6.2 , 7.9]}
d1 = pd.DataFrame(details)
d1
```

```python
d2 = d1.drop([2,3] , axis='rows')
d2
```

```python
d3 = d1.drop('Salary' , axis='columns')
d3
```

****

**Example of fillna()**
****

```python
details = {'name' : ['Abhay' , 'Vijay' , 'John' , 'Jacob' , 'Jerry'],
           'Salary' : ['$3000' , '$2545' , '$2000' , '$40000' , '$50000'],
           'exp' : [2.4 , 4.3, 5.5, np.nan , 7.9]}
d1 = pd.DataFrame(details)
d1
```

```python
d1['exp'] = d1['exp'].fillna(d1['exp'].mean())
d1
```

**Fix Missing values in age column on analysing if age id dependent on other factors**
***

```python
df.head()
```

```python
#Average age of Male and Female Passengers

df.groupby('Sex')['Age'].mean()
```

```python
df.groupby('Pclass')['Age'].mean()
```

```python
df.isnull().sum()
```

```python
#seperate data samples with missing age values and not missing age values

dfa0 = df[df['Age'].isnull()]
dfa0.head()
```

```python
dfa1 = df[~df['Age'].isnull()]
dfa1.head()
```

```python
dfa0['Age'] = dfa0['Pclass'].apply(lambda x : 25.14 if x == 3 else 29.87 if x == 2 else 38.23)
```

```python
dfa0.head()
```

```python
#concat() - to join two different data frames

df_new = pd.concat((dfa0,dfa1) , ignore_index=True)
df_new.head()
```

```python
df_new = df_new.drop('Cabin' , axis='columns')
df_new.head()
```

```python
sns.boxplot(df_new['Age'])
```

```python
df_new = df_new.dropna(axis='rows' , subset=['Embarked'])
df_new.head()
```
